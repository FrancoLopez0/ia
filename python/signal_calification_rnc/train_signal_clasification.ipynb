```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense

classifier = Sequential()
```

```python
# Agrego una capa de convolucion la cual identificara 32 rasgos (mapas de caracteristicas) con un kernels de 3x3
classifier.add(Convolution2D(filters=32, kernel_size=(3,3), input_shape=(400,400,1)))

# Agrego una capa de Max pooling la cual permite la compresion y escalado de los datos obtenidos
classifier.add(MaxPooling2D())

# Transformo las matrices en arreglos de 1 dimension para pasarlos a una RNA
classifier.add(Flatten())
```

```python
classifier.add(Dense(units = 128 , activation="relu"))

classifier.add(Dense(units = 64, activation="relu"))

classifier.add(Dense(units = 4, activation="softmax"))
```

```python
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
